```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name = "training_results"

data = pd.read_csv(file_name + ".csv")

print(data.head())

```

```python
# 데이터의 열 수를 가져와서 x축의 데이터로 사용
epochs = range(len(data))

# Accuracy 그래프 그리기
plt.plot(epochs, data['Accuracy'], label='Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy')
plt.legend()
plt.show()
```

```python
# Precision, Recall, F1 Score 그래프 그리기
plt.plot(epochs, data['Precision'], label='Precision')
plt.plot(epochs, data['Recall'], label='Recall')
plt.plot(epochs, data['F1 Score'], label='F1 Score')
plt.xlabel('Epoch')
plt.ylabel('Valid')
plt.title('Precision, Recall, F1 Score')
plt.legend()
plt.show()
```
